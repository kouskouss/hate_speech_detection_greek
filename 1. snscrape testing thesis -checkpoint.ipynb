```python
# create a list with only the convo id
link = 'https://twitter.com/Epicrotus/status/1368946528895787022'
link.split("/")[-1]

```

```python
linksoftweets = ['https://twitter.com/Epicrotus/status/1368946528895787022', 
'https://twitter.com/EFSYNTAKTON/status/1634170594823483394', 
'https://twitter.com/MsLouellaDeVil/status/1633694672664276992', 
'https://twitter.com/theodorikakosp/status/1634547017492865024', 
'https://twitter.com/theodorikakosp/status/1626905858742386690',
'https://twitter.com/ArisPortosalte/status/1633564998017449985',
'https://twitter.com/ellada24/status/1632706990236393472',
'https://twitter.com/karamanlis_k/status/1630933468170403842',
'https://twitter.com/AdonisGeorgiadi/status/1634157479859765248',
'https://twitter.com/ellada24/status/1635257393494777856',
'https://twitter.com/katrinken/status/1635361609886605314',
'https://twitter.com/katrinken/status/1635361502860546048',
'https://twitter.com/nikos_sverkos/status/1636338432803840002']

linksoftweets[1].split("/")[-1]

conversation_ids = [link.split("/")[-1] for link in linksoftweets]
```

```python
conversation_ids
```

```python
tweets_replies = []
```

```python
for n, k in enumerate(conversation_ids):
    for i,tweet in enumerate(twitter.TwitterSearchScraper(' conversation_id:'+ conversation_ids[n] +'(filter:safe OR -filter:safe) since:2021-01-01 until:2023-03-25 lang:el').get_items()):
        if i>100:
            break
        tweets_replies.append([tweet.date, tweet.user.username, tweet.content, conversation_ids[n]])
```

```python
tweets_replies
```

```python
# tweets from replies from posts
tweets_replies = pd.DataFrame(tweets_replies, columns=['Date', 'Username', 'Tweet', 'ID'])
print(tweets_replies)
```
